```python
from compsyn.vectors import WordToColorVector
from compsyn.config import CompsynConfig

#(CompsynConfig(jzazbz_array='/anaconda3/lib/python3.7/site-packages/compsyn/jzazbz_array.npy'))
```

```python
def quick_w2cv(label: str) -> WordToColorVector:
    w2cv = WordToColorVector(label=label)
    w2cv.run_image_capture()
    w2cv.run_analysis()
    return w2cv
```
